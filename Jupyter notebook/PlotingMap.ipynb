```python
x = [1,2,3,4,5]
y = [1000,2500,4700,14500,20000]
maintainance = [200,500,1500,2500,7000]
plt.plot(x,y,color = "#212121")
plt.plot(x,maintainance,color = "green")
plt.xlabel("Bed Hall Kitechen")
plt.ylabel("Rent in thousand")
plt.title("Flat Rent in our area")
plt.legend("one","two")
```

```python
bhk = [1,2,3,4,5]
rent = [1000,2500,4700,14500,20000]
maintainance = [200,500,1500,2500,7000]
plt.plot(bhk,rent,color = "#03DAC6")
plt.plot(bhk,maintainance,color = "red")

x1 = plt.subplot(111)
x1.plot(bhk,rent,label='Rent')
x1.plot(bhk,maintainance,label='Maintainance')
plt.title('This is Legend')
x1.legend()

plt.xlabel("Bed Hall Kitechen")
plt.ylabel("Rent in thousand")
plt.title("Flat Rent in our area")
```
